```python
import pandas as pd
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
code_results = pd.read_csv('699_results.csv')
code_results
```

```python
ppl_results = pd.read_csv('pick2-results.csv')
ppl_results
```

```python
def read_dict(string, output):
    if output == 1:
        entries = string[1:-1].replace("'", "").split(', ')
        dix = {}
        for line in entries:
            color, p = line.split(': ')
            dix[color] = float(p)
        
        return dix

    if output == 2:
        fg = string[1:-1].replace("'", "").split(', ')[:11]
        fg[0] = fg[0].replace('fg: {', '')
        fg[-1] = fg[-1].replace('}', '')
        bg = string[1:-1].replace("'", "").split(', ')[11:]
        bg[0] = bg[0].replace('bg: {', '')
        bg[-1] = bg[-1].replace('}', '')
        
        fg_dix = {}
        bg_dix = {}
        for line in fg:
            color, p = line.split(': ')
            fg_dix[color] = float(p)
        for line in bg:
            color, p = line.split(': ')
            bg_dix[color] = float(p)
    
        return fg_dix, bg_dix
```

```python
# takes the two colors that cover the most pixels in the full image
def strat1():
    labels_1 = []
    labels_2 = []
    for i in range(20):
        line = code_results['output1'].iloc[i]
        dix = read_dict(line, 1)
        ordered_dix = sorted(dix.items(), key=lambda i:i[1], reverse=True)
        labels_1.append(ordered_dix[0][0])
        labels_2.append(ordered_dix[1][0])
    
    code_results['strat1_label1'] = labels_1
    code_results['strat1_label2'] = labels_2
    
    return code_results
```

```python
# takes the top 2 colors that cover the most pixels in the foreground of the image
def strat2():
    labels_1 = []
    labels_2 = []
    for i in range(20):
        line = code_results['output2'].iloc[i]
        fg_dix, bg_dix = read_dict(line, 2)
        ordered_dix = sorted(fg_dix.items(), key=lambda i:i[1], reverse=True)
        labels_1.append(ordered_dix[0][0])
        labels_2.append(ordered_dix[1][0])
    
    code_results['strat2_label1'] = labels_1
    code_results['strat2_label2'] = labels_2
    
    return code_results
```

```python
#  uses the most common color of the foreground pixels and the most common color of the background pixels
def strat3():
    labels_1 = []
    labels_2 = []
    for i in range(20):
        line = code_results['output2'].iloc[i]
        fg_dix, bg_dix = read_dict(line, 2)
        
        ordered_dix1 = sorted(fg_dix.items(), key=lambda i:i[1], reverse=True)
        labels_1.append(ordered_dix1[0][0])
        
        ordered_dix2 = sorted(bg_dix.items(), key=lambda i:i[1], reverse=True)
        if ordered_dix2[0][0] == ordered_dix1[0][0]:
            labels_2.append(ordered_dix2[1][0])
        else:
            labels_2.append(ordered_dix2[0][0])
        
    
    code_results['strat3_label1'] = labels_1
    code_results['strat3_label2'] = labels_2
    
    return code_results
```

```python
# uses the 2 most common colors in the background
def strat4():
    labels_1 = []
    labels_2 = []
    for i in range(20):
        line = code_results['output2'].iloc[i]
        fg_dix, bg_dix = read_dict(line, 2)
        ordered_dix = sorted(bg_dix.items(), key=lambda i:i[1], reverse=True)
        labels_1.append(ordered_dix[0][0])
        labels_2.append(ordered_dix[1][0])
    
    code_results['strat4_label1'] = labels_1
    code_results['strat4_label2'] = labels_2
    
    return code_results
```

```python
def human_labels():

    human_label_top1 = []
    human_label_top2 = []
    human_label_all = []
    
    for img in code_results['filename']:
        count = {}
        for i in range(ppl_results.shape[0]):
            string = ppl_results[img].iloc[i]
            answers = string.lower().split(';')
            for n in answers:           
                if n in count.keys():
                    count[n] += 1
                else:
                    count[n] = 1
                    
            top = sorted(count.items(), key=lambda i:i[1], reverse=True)

        human_label_top1.append(top[0][0])
        human_label_top2.append(top[1][0])
        human_label_all.append(set(count.keys()))
    
    code_results['human_label1'] = human_label_top1
    code_results['human_label2'] = human_label_top2
    code_results['human_label_all'] = human_label_all
    
    return code_results
```

```python
strat1()
strat2()
strat3()
strat4()
human_labels()
code_results
```
